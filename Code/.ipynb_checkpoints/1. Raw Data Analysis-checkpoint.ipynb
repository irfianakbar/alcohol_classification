# 3. Data prprocessing

```python
qcm3_df['MIP_channel'] = 1
qcm6_df['MIP_channel'] = 1
qcm7_df['MIP_channel'] = 1
qcm10_df['MIP_channel'] = 1
qcm12_df['MIP_channel'] = 0

qcm3_df['NP_channel'] = 1
qcm6_df['NP_channel'] = 0
qcm7_df['NP_channel'] = 0.5
qcm10_df['NP_channel'] = 2
qcm12_df['NP_channel'] = 1

qcm3_df['MIP_NP_ratio'] = qcm3_df['MIP_channel']/qcm3_df['NP_channel']
qcm6_df['MIP_NP_ratio'] = qcm6_df['MIP_channel']/qcm6_df['NP_channel']
qcm7_df['MIP_NP_ratio'] = qcm7_df['MIP_channel']/qcm7_df['NP_channel']
qcm10_df['MIP_NP_ratio'] = qcm10_df['MIP_channel']/qcm10_df['NP_channel']
qcm12_df['MIP_NP_ratio'] = qcm12_df['MIP_channel']/qcm12_df['NP_channel']
```

```python
qcm_all_df = pd.concat([qcm3_df, qcm6_df, qcm7_df, qcm10_df, qcm12_df], ignore_index=True)
print(qcm_all_df.shape)
qcm_all_df.head()
```

```python
qcm_all_df.loc[qcm_all_df["1-Octanol"] == 1, 'label'] = '1-Octanol'
qcm_all_df.loc[qcm_all_df["1-Propanol"] == 1, 'label'] = '1-Propanol'
qcm_all_df.loc[qcm_all_df["2-Butanol"] == 1, 'label'] = '2-Butanol'
qcm_all_df.loc[qcm_all_df["2-propanol"] == 1, 'label'] = '2-propanol'
qcm_all_df.loc[qcm_all_df["1-isobutanol"] == 1, 'label'] = '1-isobutanol'
```

```python
def stat_kde_plot(df,input1,input2):
    f, (ax1,ax2) = plt.subplots(1,2,figsize=(15,5))
    sns.kdeplot(df[input1],ax = ax1,color ='blue',shade=True,
                label=("Skewness : %.2f"%(df[input1].skew()),
                       "Kurtosis: %.2f"%(df[input1].kurtosis())))
    sns.kdeplot(df[input2], ax = ax2,color='r',shade=True,
                label=("Skewness : %.2f"%(df[input2].skew()),
                       "Kurtosis: %.2f"%(df[input2].kurtosis())))
    axes = [ax1,ax2]
    input_cols = [input1,input2]
    for j in range(len(axes)):
        axes[j].set_xlabel(input_cols[j],color='black',fontsize=12)
        axes[j].set_title(input_cols[j] + ' Kdeplot',fontsize=14)
        axes[j].axvline(df[input_cols[j]].mean() , color ='g',linestyle = '--')
        axes[j].legend(loc ='upper right',fontsize=12,ncol=2)
    sns.despine()
    return plt.show()


def stat_plot (df, input_cols):
    (mu, sigma) = norm.fit(df[input_cols])
    f, (ax1, ax2)= plt.subplots(1,2,figsize=(15,5))
    
    # Apply the log transformation on the column
    sns.distplot(df[input_cols],ax = ax1,fit=norm,color ='blue',hist=False)
    ax1.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],loc='best')
    ax1.set_ylabel('Frequency')
    ax1.set_title(input_cols +' Distribution',color='black',fontsize=14)
    
    #Get also the QQ-plot
    res = stats.probplot(df[input_cols], plot=ax2)
    sns.despine()
    return plt.show()
```

```python
stat_kde_plot(qcm_all_df, '0.799_0.201_MIP', '0.799_0.201_NP')
stat_kde_plot(qcm_all_df, '0.700_0.300_MIP', '0.700_0.300_NP')
stat_kde_plot(qcm_all_df, '0.600_0.400_MIP', '0.600_0.400_NP')
stat_kde_plot(qcm_all_df, '0.501_0.499_MIP', '0.501_0.499_NP')
stat_kde_plot(qcm_all_df, '0.400_0.600_MIP', '0.400_0.600_NP')
stat_kde_plot(qcm_all_df, 'MIP_channel', 'NP_channel')
```

```python
for col in fea_cols:
    stat_plot(qcm_all_df, col)
    plt.show()
```

# 5. Train Test Splitting

```python
label_cols = ['1-Octanol', '1-Propanol', '2-Butanol', '2-propanol', '1-isobutanol', 'label']

X = qcm_all_df[fea_cols]
y = qcm_all_df[label_cols]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# 6. Scaling

```python
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```
